Используя алгоритм TD-PSOLA, замедлите сигнал cta0001.wav в два раза. Для этого:

1. Получите данные о метках периодов (из разметки или с помощью Praat)
2. Разделите сигнал на окна, получите последовательность ST-сигналов. Не забудьте о необходимости извлекать окна и из глухих участков тоже!
3. Определите, где должны быть новые метки периодов (их должно быть больше, но общая форма контура должна остаться той же самой)
4. Для каждой новой метки подберите подходящий ST-сигнал
5. Сложите всё вместе
6. Запишите в новый файл
